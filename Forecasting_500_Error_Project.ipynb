```python
df_500 = df[(df['httpStatus'] >= 500) & (df['httpStatus'] < 600)]
```

```python
df_500 = df[(df['httpStatus'] >= 500) & (df['httpStatus'] < 600)].copy()
```

```python
df_500['minute'] = df_500['_time'].dt.floor('min')
```

```python
df_500.shape
```

```python
df_500.head()
```

```python
error_counts = df_500.groupby('minute').size().reset_index(name='error_count')
```

```python
error_counts.shape
```

```python
error_counts.head()
```

```python
full_range = pd.date_range(start=df['_time'].min().floor('min'), end=df['_time'].max().ceil('min'), freq='min')
full_df = pd.DataFrame({'minute': full_range})
```

```python
error_counts['minute'] = pd.to_datetime(error_counts['minute']).dt.tz_localize(None)
full_df['minute'] = pd.to_datetime(full_df['minute']).dt.tz_localize(None)
```

```python
full_df.shape
```

```python
merged_df = full_df.merge(error_counts, on='minute', how='left').fillna(0)
merged_df['error_count'] = merged_df['error_count'].astype(int)
```

```python
from prophet import Prophet
prophet_df = merged_df.rename(columns={'minute': 'ds', 'error_count': 'y'})
prophet_df['ds'] = prophet_df['ds'].dt.tz_localize(None)
```

```python
prophet_df.head()
```

```python
model = Prophet()
model.fit(prophet_df)
```

```python
future = model.make_future_dataframe(periods=15, freq='min')
forecast = model.predict(future)
```

```python
model.plot(forecast);
```

```python
import matplotlib.pyplot as plt

model.plot(forecast)
plt.title("Forecast of 500-series Errors for Next 15 Minutes")
plt.xlabel("Time")
plt.ylabel("Predicted Error Count")
plt.show()
```
